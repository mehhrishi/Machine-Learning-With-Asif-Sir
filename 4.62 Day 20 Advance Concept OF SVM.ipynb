```python
import warnings
warnings.filterwarnings("ignore")
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
colname=["sepal length in cm","sepal width in cm","petal length in cm","petal width in cm","class"]

df = pd.read_csv(path, header=None, names=colname)
df
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y =  le.fit_transform(y)
```

```python
y
```

```python
le.classes_
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
from sklearn.svm import SVC
svm = SVC()
svm.fit(xtrain,ytrain)
ypred = svm.predict(xtest)
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest,ypred))
```

```python
train =svm.score(xtrain,ytrain)
test = svm.score(xtest,ytest)

print(f"Training Accuracy:- {train}\n Testing Accuracy:- {test}")
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
```

```python
pipe = Pipeline(
                  steps=[("scaler",StandardScaler()),
                         ("svm",SVC())
                      
                  ]
 
)
```

```python
pipe.fit(xtrain,ytrain)
ypred = pipe.predict(xtest)
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest,ypred))
```

```python
train =pipe.score(xtrain,ytrain)
test = pipe.score(xtest,ytest)

print(f"Training Accuracy:- {train}\n Testing Accuracy:- {test}")
```

# GridSearch CV

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameter = {
             "C":[0.1,1,10],
             "gamma":[0.1,0.01,0.001],
             "kernel":["rbf"]
}
```

```python
grid = GridSearchCV(SVC(), parameter, verbose=2)
grid.fit(xtrain,ytrain)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
grid.best_estimator_
```

```python
svm = grid.best_estimator_
svm.fit(xtrain,ytrain)
ypred = svm.predict(xtest)
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest,ypred))
```
