## Split the Dataset X and Y

```python
df.head()
```

```python
x = df.iloc[:,0:1]
x
```

```python
y = df.loc[:,'Price']
y
```

## Train Test Split 

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.3,
                                             random_state=1)
```

## Model Evaluation

```python
# step1:- Import the Model 
from sklearn.linear_model import LinearRegression

# step2:- Object Creation
linreg = LinearRegression()

# step3:- Train the Model using Traing Data -----> m and c (y = mx+c)
# In this step, the model first calculatye the Slope(m) value,
# with the help of xtrain and ytrain 
# then after finding the slope(m), it will calculate the intercept(c)
linreg.fit(xtrain,ytrain)
# in the above query the Model is ggetting trained using Training Dataset 

# Step4:- Make Prediction 
# Now lets try and pass "xtest" which will act as a New Observation 
# where "xtest" already has an Actual output as "ytest"
ypred = linreg.predict(xtest)
```

## Regression Evaluation Metrics

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score

mae = mean_absolute_error(ytest,ypred)

mse = mean_squared_error(ytest,ypred)

rmse = np.sqrt(mse)

r2 = r2_score(ytest,ypred)

print(f" MAE :- {mae}\n MSE:- {mse}\n RMSE:- {rmse}\n ACCURACY:- {r2}")
```

```python
def predict(x,y):
    from sklearn.model_selection import train_test_split
    xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.3,
                                             random_state=1)
    # step1:- Import the Model
    from sklearn.linear_model import LinearRegression
    
    # step2:- Object Creation
    linreg = LinearRegression()
    
    # step3:- Train the Model using Training Data ----> m and c
    linreg.fit(xtrain,ytrain)
    
    # step4:- Make Prediction
    ypred = linreg.predict(xtest)
    
    print('Co-efficient: ',linreg.coef_)
    print('Intercept: ',linreg.intercept_)
    
    return ypred
    
x=list(df.select_dtypes(float).columns)
x=x[:-1]
for i in x:
    print(i)
    a=predict(df[[i]],df['Price'])
    print(a)
    print("Lets plot the graph")
    # Actual value of xtrain which is "ytrain"
    plt.scatter(xtrain,ytrain,color="red")
    plt.plot(xtrain,a,color='blue')
    plt.show()
    print('\n')
# print(ypred)
```
