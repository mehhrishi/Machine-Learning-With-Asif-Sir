# MinMaxScaler

```python
# xnew = (x-min)/(max-min)
```

```python
a = np.array([1,2,3,4,5])
b = np.array([135,202,3032,4032,3250])
c = np.array([100,200,300,400,500])
```

```python
def minmax(arr):
    xnew = (arr - arr.min())/(arr.max()-arr.min())
    return xnew
```

```python
minmax(a)
```

```python
minmax(b)
```

```python
minmax(c)
```
